```python
# List all files in a directory
path = '/content/drive/MyDrive/BRIN/Segmentation-datasets/segmented'
files = []
for f in os.listdir(path):
    if os.path.isfile(os.path.join(path, f)):
        files.append(f)
```

```python
#print(path)
#for i in files[:1000]:
#  print(i)
```

```python
os.makedirs('/content/drive/MyDrive/BRIN/Segmentation-datasets/mask_baru')
```

```python
new_path = '/content/drive/MyDrive/BRIN/Segmentation-datasets/mask_baru/'
for file in files:
  image_mask = cv2.imread(path + '/' + file, cv2.IMREAD_GRAYSCALE)
  image_mask[image_mask!=0]=1
  np.save(new_path + file[:-4] +' .npy', image_mask)
```
