```python
SIZE = 256
num_images = 1000
```

```python
image_names = glob.glob("/kaggle/working/color/*")
image_names.sort()
image_names_subset = image_names[0:num_images]
images = [cv2.imread(img) for img in image_names_subset]
#gray_scale = rgb2gray(images)
image_dataset = np.array(images)
#image_dataset = np.expand_dims(image_dataset, axis = 3)
```

```python
mask_names = glob.glob("/kaggle/working/mask_baru/*.npy")
mask_names.sort()
mask_names_subset = mask_names[0:num_images]
masks = [np.load(mask) for mask in mask_names_subset]
mask_dataset = np.array(masks)
mask_dataset = np.expand_dims(mask_dataset, axis = 3)
```

```python
print("Image data shape is: ", image_dataset.shape)
print("Mask data shape is: ", mask_dataset.shape)
print("Max pixel value in image is: ", image_dataset.max())
# print("Labels in the mask are : ", np.unique(mask_dataset))
```

```python
#Normalize images
image_dataset = image_dataset /255.  #Can also normalize or scale using MinMax scaler
#Do not normalize m atmasks, just rescale to 0 to 1.
# mask_dataset = mask_dataset /255.  #PIxel values will be 0 or 1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(image_dataset, mask_dataset, test_size = 0.20, random_state = 42)
```

```python
!pip install transunet
```

```python
from transunet import TransUNet

# Create the model
model = TransUNet(image_size=256, pretrain=True, num_classes = 1)

# Summary
#model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate = 1e-3), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train,
                    batch_size = 16,
                    verbose=1,
                    epochs=25,
                    validation_data=(X_test, y_test),
                    shuffle=False)
```

```python
data_dir = glob.glob("/kaggle/input/train-test-pd-2/train-test PD/Tomato___healthy/*")
```

```python
def apply_mask_to_image(original_image, mask_image):
    original_image = original_image
    new_image = np.zeros_like(original_image)
    for i in range(mask_image.shape[0]):
        for j in range(mask_image.shape[1]):
            if mask_image[i, j] == 1:
                new_image[i, j, :] = original_image[i, j, :]

    return new_image
```

```python
save_dir = '/kaggle/working/Hasil_TransUnet_PD/Tomato___healthy'
for img in data_dir:
  image = cv2.imread(img)
  test = np.squeeze(cv2.resize(image, (256,256)))
  norm = test/255
  expand = np.expand_dims(norm, axis = 0)
  pred = np.squeeze((model.predict(expand)[0,:,:,0] > 0.5))
  a = apply_mask_to_image(test, pred)
  name = img.split('/')[-1]
  cv2.imwrite(save_dir + name, a)
```
